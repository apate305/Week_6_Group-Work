2. Create a visualization that shows the top 5 categories (those referenced the most often). Be sure to label appropriately.

```python
categories = jeo_df[" Category"].value_counts()
categories_df = pd.DataFrame(categories)
top_5_categories = categories_df.head()

#Convert index to column by adding new index
top_5_categories.reset_index(level=0, inplace=True)
top_5_categories

#BEFORE & AFTER, SCIENCE, LITERATURE, AMERICAN HISTORY, POTPOURRI  
```

```python
import pandas as pd

import matplotlib.pyplot as plot

#'rot' allows the user to shift the xlabels so that the labels are not on top of each other--'rotate'

top_5_categories.plot.bar(x= "index", y=" Category", rot=20, title="Top 5 Categories");

plot.show
```
